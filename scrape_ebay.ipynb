```python
def find_products(item):
    
    names, prices, urls = [], [], []
    p = 1
    
    while(p > 0):
        url = f'https://www.ebay.com/sch/i.html?_from=R40&_nkw={item}&_sacat=0&LH_PrefLoc=1&_ipg=240&_pgn={p}'
        dr.get(url)
        name_list = dr.find_elements(By.CLASS_NAME, 's-item__title')
        price_list = dr.find_elements(By.CLASS_NAME, 's-item__price')
        url_list = dr.find_elements(By.CLASS_NAME, 's-item__link')

        if (len(url_list)) > 0 :
            for i in range(len(url_list)):
                try:
                    url_list[i] = url_list[i].get_attribute('href')
                except:
                    url_list[i] = 'NAN'
                try:
                    name_list[i] = name_list[i].get_attribute('innerText')
                except:
                    name_list[i] = 'NAN'
                try:
                    price_list[i] = price_list[i].get_attribute('innerText')
                except:
                    price_list[i] = 'NAN'
                
            names.extend(name_list)
            prices.extend(price_list)
            urls.extend(url_list)
            p += 1
            sleep(5)
            
        else:
            p = 0
    
    il = pd.DataFrame(data=[names, prices, urls]).transpose()
    il.columns = ['name','price', 'url']
    return il
```

```python
item_list = find_products(item)
```
